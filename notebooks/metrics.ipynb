```python
import pandas as pd

df_ensemble = pd.read_csv('../samples/mean_metric_ensemble.csv', header=None, names=['Metric', 'Value'])
df_multiquery = pd.read_csv('../samples/mean_metrics.csv', header=None, names=['Metric', 'Value'])

df_ensemble['Retriever'] = 'Ensemble Retriever'
df_multiquery['Retriever'] = 'Multiquery Retriever'

df_combined = pd.concat([df_ensemble, df_multiquery], ignore_index=True)
df_cleaned = df_combined.dropna(subset=['Metric'])
df_cleaned.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))

sns.barplot(data=df_cleaned, x='Metric', y='Value', hue='Retriever', palette='Paired')

plt.title('Comparison of RAGAs Metrics for Retrievers')
plt.xlabel('Metric')
plt.ylabel('Value')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.lineplot(data=df_combined, x='Metric', y='Value', hue='Retriever', marker='o', palette='viridis')

plt.title('RAGAs Metrics Comparison for Retrievers')
plt.xlabel('Metric')
plt.ylabel('Value')

plt.grid(True)
plt.show()
```
